**Estructuras de datos:**


Listas: Puede contener elementos de diferentes tipos. Podemos modificar la lista asignando nuevos elementos, se dice que la lista es mutable




```python
#Creando listas
l = [1,2,5.4,4]

print(type(l))
print(l)
```

```python
# Crear una lista vacia
l = []

# Agregando elementos utilizando `append`

l.append(10)
l.append(20)
l.append(30)

print(l)
```

```python
# Asignando nuevos valores a la lista
l[1] = 45
l[2] = 55
print(l)
```

```python
l.insert(0, 22)
l.insert(0, 32)
l.insert(0, 42)
l.insert(0, 52)

print(l)
```

```python
# Removemos un elemento con un valor específico
l.remove(45)

print(l)
```

```python
#Removiendo de una locación especifica
del l[0]
del l[3]

print(l)
```

Tuples: Se parecen a las listas, pero no pueden ser modificadas una vez creadas  Se crean utilizando la sintaxis (...,...,..) ó ...,...
son inmutables.

```python
point = (10, 2.0)

print(point, type(point))
```

```python
point = 10, 2.0

print(point, type(point))
```

```python
x, y = point

print("x =", x)
print("y =", y)
```

```python
#Si asignamos un nuevo valor obtendremos error.
#point[0] = 20
```

Diccionarios: Parecidos a las listas, pero cada leemento es un par de key-values. La sintaxis es {key1: value1, ....}

```python
params = {"parameter1" : 2.0,
          "parameter2" : 7.0,
          "parameter3" : 18.0,}

print(type(params))
print(params)
```

```python
params["parameter1"] = "80.0"
params["parameter2"] = "C"

# add a new entry
params["parameter4"] = "75"

print("parameter1 = " + str(params["parameter1"]))
print("parameter2 = " + str(params["parameter2"]))
print("parameter3 = " + str(params["parameter3"]))
print("parameter4 = " + str(params["parameter4"]))
```

Creando listas utilizando "for loops"

```python
l1 = []
for x in range(0, 100):
    if x % 2 == 0:
        l1.append(x**2)
print(l1)
```

```python
l1 = [x**2 for x in range(0,100) if x % 2 == 0]
print(l1)
```

#**Numpy**

Este paquete agrega  funcionalidades  para trabajar con vectores, matrices y estructuras de más dimensiones. La terminología utilizada para estos conjuntos de datos es numpy.ndarray.



*   Permite especificar el tipo de datos que queremos utilizar (Ejemplo: int8, uint32, float64)

*   Funciones de manipulacion de arreglos (Ejemplo: reshape, transpose, concatenate)

*   Generación de arreglos (Ejemplo: ones, zeros, eye, random)

*   Operaciones matemáticas elemento por elemento (Ejemplo: add, multiply, max, sin)

*   Operaciones matemáticas de matrices (Ejemplo: add, multiply, max, sin)

*   Álgebra lineal (Ejemplo: inv, pinv, svd, eig, det, qr)



```python
#Importando libreria
import numpy as np
print("NumPy importado con éxito.")
```

```python
# Vector: el argumento de la función de array es una lista de Python.
v = np.array([10,2,3,8,-5])
print("vector:",v)
```

```python
#Matriz: el argumento de la función de array es una lista de Python anidada.
M = np.array([[1, 2], [3, 4]])
M
```

```python
type(v), type(M)
```

**Propiedades de ndarray**



```python
# ndarray.shape: Nos devuelve una tuple con el número de elementos en cada eje
print("Dimensiones del vector:", v.shape)
print("Dimensiones de la matriz:", M.shape)

# ndarray.ndim: Nos devuelve las dimensiones del arreglo
print("Número de dimensiones del vector:", v.ndim)
print("Número de dimensiones de la matriz:", M.ndim)
```

**Manipulación de arreglos**

Indexación

```python
print("vector:",v)
print("Primer elemento de v: ", v[0])
print("Elementos de la posición 1 y 2: ", v[1:3]) # v[lower:upper:step]
print("Extracción de lementos con step = 2: ", v[::2]) # v[lower:upper:step]
print("Primeros 3 elemntos de v: ", v[:3])
print("Último elemento de v: ", v[-1])
print("Últimos 3 elementos de v: ", v[-3:])
```

```python
print("Matriz:", M)
print("Elemento de la fila 0, y columna 1 de la matriz M: ", M[0][1])
print("Primera fila de la matriz: ", M[0,:])
print("Segunda columna de la matriz: ", M[:,1])
```

**Operaciones con arrays**

```python
a = np.array([1, 2, 3])
print("Array a: ", a)

b = np.array([1, 2, 3])

c = np.array([1, 2, 3])
print("Array c: ", c)


d = 2 * a
print("Multiplicación escalar de a por 2: ", d)

c = a + b
print("Suma elemento por elemento de arrays a y c: \n", c)

e = np.mean(a)
print("Media de los elementos de a: ", e)

max_val = np.max(a)
print("Valor máximo de a: ", max_val)

min_val = np.min(a)
print("Valor mínimo de a: ", min_val)

sum_val = np.sum(a)
print("Suma de los elementos de a: ", sum_val)
```

```python
c = np.multiply(a, b)
print("Multiplicación elemento a elemento de a y b: \n", c)

d = np.dot(a, b)
print("Producto punto de a y b: \n", d)

e = np.cross(a, b)
print("Producto cruz de a y b: \n", e)

f = a @ b
print("Multiplicación matricial de a y b: \n", f)
```

**Uso de funciones generadoras de arreglos**

```python
# np.arange()
arr1 = np.arange(0, 10, 2) # argumentos: inicio, fin, step
print("Vector de 0 a 10  de 2 en 2 ", arr1)

# np.zeros(), np.ones
arr2 = np.zeros(5)
print("Vector de ceros con 5 elementos", arr2)

# np.linspace()
arr3 = np.linspace(0, 10, num=5) # El último punto es incluido
print("Vector de 5 elementos igualmente espaciados de 0 a 10", arr3)

# mp.identity
arr4 = np.identity(3)
print("Matriz identidad de 3x3", arr4)
```

**Normas de vectores**

```python
v = np.array([1, 2, 3, 4, 5])
norm1 = np.linalg.norm(v, ord=1)
print("Norma 1 de v: ", norm1)

norm2 = np.linalg.norm(v, ord=2)
print("Norma 2 de v: ", norm2)

norm_inf = np.linalg.norm(v, ord=np.inf)
print("Norma infinita de v: ", norm_inf)
```

#**Matplotlib**

Este paquete nos permite generar gráficos de 2D y 3D.

```python
import matplotlib.pyplot as plt
import numpy as np
```

Ejemplo de gráfica de línea

```python
x = np.linspace(0, 2*np.pi*5, 1000) # Generamos 1000 valores entre 0 y 10pi
y = np.sin(x)
plt.plot(x, y)
plt.title("Gráfico de línea")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

Cambiar tamño de fuente

```python
plt.plot(x, y, label="5 ciclos")
plt.title("Título", fontsize=20)
plt.legend(fontsize=14)
plt.legend(loc='lower right', fontsize= 14)
plt.xlabel("x", fontsize=16)
plt.ylabel("sin(x)", fontsize=16)
```

Cambiar colores

```python
plt.plot(x, y, color="r")
```

```python
plt.plot(x, y, color="g")
```

```python
plt.plot(x, y, color="b")
plt.scatter(x, y, c=x, cmap="coolwarm")
```

Tipos de gráficos
Ejemplo de gráfico de barras


```python
x = ["A", "B", "C", "D", "E"]
y = [3, 6, 2, 8, 4]
plt.bar(x, y)
plt.title("Gráfico de barras")
plt.xlabel("Categorías")
plt.ylabel("Valores")
plt.show()
```

Ejemplo de gráfico de dispersión

```python
x = np.random.randn(100)
y = np.random.randn(100)
plt.scatter(x, y)
plt.title("Gráfico de dispersión")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.show()
```

Ejemplo de gráfico de áreas

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.fill_between(x, y1, y2, alpha=0.5)
plt.title("Gráfico de áreas")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.show()
```

Multiples gráficos en una imagen

```python
x = np.linspace(0, 2*np.pi*1.5, 100)
y = np.sin(x)

## Comando adicional para cambiar el tamano de los graficos en el notebook
plt.figure(figsize=(18,10))
plt.plot(x, 2*y, '-o', label="Mi gráfico 1", linewidth=2)
plt.plot(x, y, '-o', label="Mi gráfico 2", linewidth=2)
plt.plot(x, y*0.5, '-o', label="Mi gráfico 3", linewidth=2)

plt.title("Graficos Multiples", fontsize=30)
plt.legend(fontsize=24)
plt.xlabel("Eje x", fontsize=26)
plt.ylabel("Eje y", fontsize=26)
plt.show()
```
